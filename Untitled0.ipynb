2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF) .


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
# Let's consider a fair six-sided die roll
low = 1  # Minimum value (inclusive)
high = 6 # Maximum value (inclusive)

# Create an array of possible outcomes (the random variable values)
outcomes = np.arange(low, high + 1)

# Calculate the probability for each outcome
# In a discrete uniform distribution, each outcome has an equal probability
probability = 1 / len(outcomes)
probabilities = np.full(len(outcomes), probability)

# Plot the Probability Mass Function (PMF)
plt.figure(figsize=(8, 5))
plt.bar(outcomes, probabilities, width=0.8, color='skyblue', edgecolor='black')
plt.xlabel('Outcome (x)')
plt.ylabel('Probability P(x)')
plt.title(f'Probability Mass Function (PMF) of a Discrete Uniform Distribution ({low} to {high})')
plt.xticks(outcomes)
plt.ylim(0, 1) # Probability ranges from 0 to 1
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution ?


```python
def bernoulli_pdf(p, x):
    """
    Calculates the Probability Distribution Function (PDF) of a Bernoulli distribution.

    Args:
        p (float): The probability of success (between 0 and 1).
        x (int): The outcome (0 for failure, 1 for success).

    Returns:
        float: The probability of the given outcome.
    """
    if not (0 <= p <= 1):
        raise ValueError("Probability 'p' must be between 0 and 1.")
    if x not in [0, 1]:
        raise ValueError("Outcome 'x' must be 0 or 1 for a Bernoulli distribution.")

    if x == 1:
        return p
    elif x == 0:
        return 1 - p

# Example usage:
probability_of_success = 0.7
outcome_success = 1
outcome_failure = 0

prob_success = bernoulli_pdf(probability_of_success, outcome_success)
prob_failure = bernoulli_pdf(probability_of_success, outcome_failure)

print(f"Probability of success (x=1) with p={probability_of_success}: {prob_success}")
print(f"Probability of failure (x=0) with p={probability_of_success}: {prob_failure}")

# Example with invalid input
try:
    bernoulli_pdf(0.5, 2)
except ValueError as e:
    print(f"Error: {e}")
```

6.  Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Reusing parameters from the previous discrete uniform distribution example
low = 1  # Minimum value (inclusive)
high = 6 # Maximum value (inclusive)
outcomes = np.arange(low, high + 1)
probability = 1 / len(outcomes)
probabilities = np.full(len(outcomes), probability)

# Calculate the Cumulative Distribution Function (CDF)
cdf = np.cumsum(probabilities)

# Plot the Cumulative Distribution Function (CDF)
plt.figure(figsize=(10, 6))
plt.step(outcomes, cdf, where='post', color='darkorange', linestyle='-', marker='o', markersize=6, label='CDF')
plt.xlabel('Outcome (x)')
plt.ylabel('Cumulative Probability F(x)')
plt.title(f'Cumulative Distribution Function (CDF) of a Discrete Uniform Distribution ({low} to {high})')
plt.xticks(outcomes)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.ylim(0, 1.05)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

9. Write a Python function to calculate Z-scores from a dataset and plot them ?


```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data, title="Distribution of Z-Scores"):
    """
    Calculates Z-scores for a dataset and plots their distribution.

    Args:
        data (array-like): The input dataset (e.g., list, numpy array).
        title (str): The title for the plot.
    """
    data = np.array(data)

    if len(data) == 0:
        print("Dataset is empty. Cannot calculate Z-scores.")
        return

    mean = np.mean(data)
    std_dev = np.std(data)

    if std_dev == 0:
        print("Standard deviation is zero. Cannot calculate Z-scores (all values are the same).")
        # You might choose to plot a single bar or handle differently if all values are identical
        return

    z_scores = (data - mean) / std_dev

    print(f"Original Data Mean: {mean:.2f}")
    print(f"Original Data Standard Deviation: {std_dev:.2f}")
    print(f"Calculated Z-scores (first 5): {z_scores[:5]}")

    # Plotting the Z-scores
    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=30, density=True, color='darkblue', edgecolor='black', alpha=0.7)
    plt.title(title)
    plt.xlabel('Z-Score')
    plt.ylabel('Density')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=1, label='Mean (Z=0)')
    plt.legend()
    plt.show()

# Example Usage with a sample dataset (you can replace this with your actual data)
# Using the normal_samples generated previously
if 'normal_samples' in locals():
    print("Using previously generated normal_samples for demonstration.")
    calculate_and_plot_zscores(normal_samples, title="Z-Scores of Normal Distribution")
else:
    print("Generating a new sample dataset for demonstration.")
    sample_data = np.random.normal(loc=50, scale=10, size=1000)
    calculate_and_plot_zscores(sample_data, title="Z-Scores of Sample Data")

# Example with a different distribution
print("\nExample with a different distribution (uniform data):")
uniform_data = np.random.uniform(0, 100, 1000)
calculate_and_plot_zscores(uniform_data, title="Z-Scores of Uniform Distribution")
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Choose a non-normal distribution: Uniform Distribution
# Parameters for the original uniform distribution
uniform_low = 0
uniform_high = 100

# 2. Define parameters for CLT simulation
num_experiments = 10000  # Number of times we take a sample and calculate its mean
sample_size = 30       # Size of each sample (typically >= 30 for CLT to hold well)

# 3. Simulate and collect sample means
sample_means = []
for _ in range(num_experiments):
    # Take a sample from the uniform distribution
    sample = np.random.uniform(uniform_low, uniform_high, sample_size)
    # Calculate the mean of the sample
    sample_means.append(np.mean(sample))

# 4. Plot the histogram of sample means
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=50, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title(f'Distribution of Sample Means (CLT Demonstration)\nOriginal: Uniform({uniform_low}, {uniform_high}), Sample Size: {sample_size}, Experiments: {num_experiments}')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Overlay a normal distribution for comparison
mean_of_sample_means = np.mean(sample_means)
std_dev_of_sample_means = np.std(sample_means)

x = np.linspace(min(sample_means), max(sample_means), 100)
y = (1 / (std_dev_of_sample_means * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean_of_sample_means) / std_dev_of_sample_means)**2)
plt.plot(x, y, color='red', linestyle='--', linewidth=2, label='Normal Distribution Fit')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Choose the original distribution: Normal Distribution
# Parameters for the original normal distribution
original_mu = 50     # Population mean
original_sigma = 10  # Population standard deviation

# 2. Define parameters for CLT simulation
num_experiments = 10000  # Number of times we take a sample and calculate its mean
sample_size = 30       # Size of each sample (typically >= 30 for CLT to hold well)

# 3. Simulate and collect sample means
sample_means_normal_dist = []
for _ in range(num_experiments):
    # Take a sample from the original normal distribution
    sample = np.random.normal(original_mu, original_sigma, sample_size)
    # Calculate the mean of the sample
    sample_means_normal_dist.append(np.mean(sample))

# 4. Plot the histogram of sample means
plt.figure(figsize=(10, 6))
plt.hist(sample_means_normal_dist, bins=50, density=True, color='purple', edgecolor='black', alpha=0.7)
plt.title(f'Distribution of Sample Means from Normal Distribution\nOriginal: N(={original_mu}, ={original_sigma}), Sample Size: {sample_size}, Experiments: {num_experiments}')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Overlay a theoretical normal distribution for comparison
mean_of_sample_means = np.mean(sample_means_normal_dist)
# Standard error of the mean (theoretical sigma for the sampling distribution)
theoretical_std_err = original_sigma / np.sqrt(sample_size)

x_theory = np.linspace(min(sample_means_normal_dist), max(sample_means_normal_dist), 100)
y_theory = (1 / (theoretical_std_err * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_theory - original_mu) / theoretical_std_err)**2)
plt.plot(x_theory, y_theory, color='red', linestyle='--', linewidth=2, label='Theoretical Normal Distribution Fit')

plt.legend()
plt.show()
```

16. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():
    """
    Calculates and plots the Probability Density Function (PDF)
    of a standard normal distribution (mean = 0, standard deviation = 1).
    """
    # Define the range for x-values (typically a few standard deviations from the mean)
    x = np.linspace(-4, 4, 1000)

    # Calculate the PDF for the standard normal distribution
    # For standard normal, loc (mean) = 0 and scale (std dev) = 1
    pdf_values = norm.pdf(x, loc=0, scale=1)

    # Plot the PDF
    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf_values, color='blue', linewidth=2)
    plt.title('Standard Normal Distribution (=0, =1)')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=1, label='Mean (=0)')
    plt.legend()
    plt.show()

# Call the function to plot the standard normal distribution
plot_standard_normal_distribution()
```

18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, dataset):
    """
    Calculates the Z-score for a given data point within a dataset.

    Args:
        data_point (float): The individual data point for which to calculate the Z-score.
        dataset (array-like): The dataset (e.g., list, numpy array) from which the data point comes.

    Returns:
        float: The calculated Z-score.
    """
    dataset = np.array(dataset)
    mean = np.mean(dataset)
    std_dev = np.std(dataset)

    if std_dev == 0:
        raise ValueError("Standard deviation is zero; cannot calculate Z-score.")

    z_score = (data_point - mean) / std_dev
    return z_score

def plot_z_score_on_normal(z_score, data_point=None, dataset_mean=None, dataset_std=None):
    """
    Plots the standard normal distribution and highlights a given Z-score.

    Args:
        z_score (float): The Z-score to highlight.
        data_point (float, optional): The original data point for display in the title.
        dataset_mean (float, optional): The mean of the original dataset for display.
        dataset_std (float, optional): The standard deviation of the original dataset for display.
    """
    x = np.linspace(-4, 4, 1000)
    pdf_values = norm.pdf(x, loc=0, scale=1)

    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf_values, color='blue', linewidth=2, label='Standard Normal PDF')
    plt.axvline(x=z_score, color='red', linestyle='--', linewidth=2, label=f'Calculated Z-score = {z_score:.2f}')
    plt.fill_between(x, 0, pdf_values, where=(x <= z_score), color='skyblue', alpha=0.3, label='Area to the left (P(Z <= z))')

    # Add details to the title
    title_str = 'Standard Normal Distribution with Highlighted Z-score'
    if data_point is not None and dataset_mean is not None and dataset_std is not None:
        title_str += f'\n(Data Point: {data_point:.2f}, Mean: {dataset_mean:.2f}, Std Dev: {dataset_std:.2f})'

    plt.title(title_str)
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.axvline(0, color='gray', linestyle=':', linewidth=1, label='Mean (Z=0)')
    plt.legend()
    plt.show()

# --- Example Usage ---

# 1. Define a sample dataset
# Using the 'normal_samples' generated earlier, or a new one if not available
if 'normal_samples' in locals():
    sample_data = normal_samples
    print("Using previously generated 'normal_samples' for demonstration.")
else:
    sample_data = np.random.normal(loc=100, scale=15, size=1000) # Mean 100, Std Dev 15
    print("Generating a new sample dataset for demonstration (Mean=100, Std Dev=15).")

# 2. Choose a data point from the dataset or define a new one
data_point_of_interest = sample_data[np.random.randint(0, len(sample_data)-1)] # Pick a random point
# Or define a specific data point, e.g.:
# data_point_of_interest = 120

# Calculate dataset statistics
dataset_mean = np.mean(sample_data)
dataset_std = np.std(sample_data)

# 3. Calculate the Z-score for the data point
try:
    calculated_z = calculate_z_score(data_point_of_interest, sample_data)
    print(f"\nOriginal Data Point: {data_point_of_interest:.2f}")
    print(f"Dataset Mean: {dataset_mean:.2f}")
    print(f"Dataset Standard Deviation: {dataset_std:.2f}")
    print(f"Calculated Z-score: {calculated_z:.2f}")

    # 4. Plot the Z-score on the standard normal distribution
    plot_z_score_on_normal(calculated_z, data_point_of_interest, dataset_mean, dataset_std)

except ValueError as e:
    print(f"Error: {e}")
```

19. Implement hypothesis testing using Z-statistics for a sample dataset ?


```python
import numpy as np
from scipy import stats

# --- Define the scenario and data ---
# Let's say we have a known population mean (0) and standard deviation ()
# For example, the average height of adult males in a country is 175 cm with a standard deviation of 7 cm.
population_mean = 175  # 0
population_std_dev = 7 # 

# We take a sample from a specific region and want to test if their average height is different.
# Sample data (hypothetical)
sample_data = np.array([178, 172, 180, 176, 174, 181, 173, 179, 177, 182, 175, 170, 177, 176, 178])
sample_size = len(sample_data)
sample_mean = np.mean(sample_data)

print(f"Hypothesized Population Mean (0): {population_mean} cm")
print(f"Known Population Standard Deviation (): {population_std_dev} cm")
print(f"Sample Size (n): {sample_size}")
print(f"Sample Mean (x): {sample_mean:.2f} cm\n")

# --- Step 1: State the Hypotheses ---
# Null Hypothesis (H0): The sample mean is equal to the population mean ( = 0)
# Alternative Hypothesis (H1): The sample mean is not equal to the population mean (  0) (Two-tailed test)
print("Hypotheses:")
print(f"H0:  = {population_mean}")
print(f"H1:   {population_mean}\n")

# --- Step 2: Choose a Significance Level () ---
significance_level = 0.05 # Commonly used 5% alpha level
print(f"Significance Level (): {significance_level}\n")

# --- Step 3: Calculate the Z-statistic ---
# Formula: Z = (x - 0) / ( / sqrt(n))
standard_error = population_std_dev / np.sqrt(sample_size)
z_statistic = (sample_mean - population_mean) / standard_error

print(f"Standard Error (/n): {standard_error:.2f}")
print(f"Calculated Z-statistic: {z_statistic:.2f}\n")

# --- Step 4: Determine the Critical Value or p-value (using p-value approach) ---
# For a two-tailed test, we need to consider both tails.
# The p-value is the probability of observing a Z-statistic as extreme as, or more extreme than, the calculated Z-statistic,
# assuming the null hypothesis is true.
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"P-value: {p_value:.4f}\n")

# --- Step 5: Make a Decision ---
print("Decision:")
if p_value < significance_level:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({significance_level}), we reject the null hypothesis.")
    print("Conclusion: There is sufficient evidence to conclude that the sample mean is significantly different from the population mean.")
elif p_value >= significance_level:
    print(f"Since the p-value ({p_value:.4f}) is greater than or equal to the significance level ({significance_level}), we fail to reject the null hypothesis.")
    print("Conclusion: There is not enough evidence to conclude that the sample mean is significantly different from the population mean.")
```

```python
import numpy as np
from scipy import stats

def calculate_confidence_interval(data, confidence_level=0.95, known_population_std=None):
    """
    Calculates a confidence interval for the mean of a dataset.

    Args:
        data (array-like): The input dataset (e.g., list, numpy array).
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).
        known_population_std (float, optional): The known population standard deviation.
                                                If None, the sample standard deviation is used,
                                                and a t-distribution is technically more appropriate
                                                for small samples, but for large samples, Z-score is a good approximation.

    Returns:
        tuple: A tuple containing (lower_bound, upper_bound).
    """
    data = np.array(data)
    n = len(data)
    if n == 0:
        raise ValueError("Dataset cannot be empty.")

    sample_mean = np.mean(data)

    # Determine the standard deviation to use
    if known_population_std:
        std_dev_for_se = known_population_std
        # For a large sample (n >= 30) or known population std, use Z-distribution
        # Using ppf (percent point function) which is the inverse of cdf
        z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
        margin_of_error = z_critical * (std_dev_for_se / np.sqrt(n))
    else:
        # If population std is unknown, use sample std (s) and t-distribution
        # For simplicity and large sample size assumption (as in previous CLT examples),
        # we will approximate with Z-distribution if n is large, otherwise t-distribution.
        # Here, we'll use t-distribution if known_population_std is not provided.
        sample_std = np.std(data, ddof=1) # ddof=1 for sample standard deviation
        if n < 30: # Small sample, use t-distribution
            t_critical = stats.t.ppf(1 - (1 - confidence_level) / 2, df=n-1)
            margin_of_error = t_critical * (sample_std / np.sqrt(n))
        else: # Large sample, Z-distribution is a good approximation for t-distribution
            z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
            margin_of_error = z_critical * (sample_std / np.sqrt(n))


    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean, margin_of_error

# --- Example Usage ---

# Using the 'normal_samples' generated earlier, or a new one if not available
if 'normal_samples' in locals():
    sample_data = normal_samples
    print("Using previously generated 'normal_samples' for demonstration.")
else:
    sample_data = np.random.normal(loc=100, scale=15, size=1000) # Mean 100, Std Dev 15
    print("Generating a new sample dataset for demonstration (Mean=100, Std Dev=15).")

confidence_level = 0.95

try:
    lower_bound, upper_bound, sample_mean, moe = calculate_confidence_interval(sample_data, confidence_level=confidence_level)

    print(f"\nSample Mean: {sample_mean:.2f}")
    print(f"Margin of Error: {moe:.2f}")
    print(f"{int(confidence_level*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    print(f"\nInterpretation: We are {int(confidence_level*100)}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")

    # Example with known population standard deviation (if available)
    print("\n--- Example with Known Population Standard Deviation ---")
    # Let's assume we know the population standard deviation is 1 for the 'normal_samples'
    if 'normal_samples' in locals():
        known_std = 1 # Assuming the original distribution had std dev of 1
        lower_bound_k, upper_bound_k, sample_mean_k, moe_k = calculate_confidence_interval(sample_data, confidence_level=confidence_level, known_population_std=known_std)
        print(f"Sample Mean (with known pop std): {sample_mean_k:.2f}")
        print(f"Margin of Error (with known pop std): {moe_k:.2f}")
        print(f"{int(confidence_level*100)}% Confidence Interval (with known pop std): ({lower_bound_k:.2f}, {upper_bound_k:.2f})")
        print(f"Interpretation: We are {int(confidence_level*100)}% confident that the true population mean lies between {lower_bound_k:.2f} and {upper_bound_k:.2f} (assuming known population std of {known_std}).")

except ValueError as e:
    print(f"Error: {e}")
```
